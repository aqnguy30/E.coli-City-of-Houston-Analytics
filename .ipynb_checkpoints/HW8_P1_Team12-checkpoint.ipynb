```python
# Load the following libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# read-in and get data ready
data = pd.read_excel('Ecoli Data.xlsx')
# want to display all the columns
pd.options.display.max_columns = 34
data.head()
```

```python
data.shape  # we know that the data has 3565 rows and 34 columns total
```

```python
data["E.coli"].isnull().sum() # we find column "E.coli" has 345 missing values 
```

```python
data['E.coli'].dropna(how='all') # drop the missing values and left with 3220 total rows in the data set
```

```python
# (1) Compute average amount of E.coli per year in Houston waterways. 
# Plot a bar graph in which x-axis is years (2013-2016) and y-axis is E.coli amount. [25 points] 

# extract the year from string format date
data['year'] = pd.DatetimeIndex(data['Date Collected']).year
data.head()

```

```python
# take only the rows of that year: 2013, 2014, 2015, 2016
date_2013 = data.loc[data['year'] == 2013]; date_2013['E.coli']
date_2014 = data.loc[data['year'] == 2014]; date_2014['E.coli']
date_2015 = data.loc[data['year'] == 2015]; date_2015['E.coli']
date_2016 = data.loc[data['year'] == 2016]; date_2016['E.coli']
```

```python
# average amount of E.coli per year: 2013,2014,2015,2016
mean_2013= date_2013['E.coli'].mean()
mean_2014= date_2014['E.coli'].mean()
mean_2015= date_2015['E.coli'].mean()
mean_2016= date_2016['E.coli'].mean()
print('The average amount of E.coli per year in Houston waterways from 2013 to 2016 \
respectively are:', round(mean_2013, 3), round(mean_2014,3), round(mean_2015,3), round(mean_2016,3))
```

```python
# Plot a bar graph in which x-axis is years (2013-2016) and y-axis is E.coli amount. 

x = ("2013","2014","2015","2016")
y = [mean_2013,mean_2014,mean_2015,mean_2016]
plt.bar(x,y)
```

```python
# (2) Compute the average amount of E.coli per month in Houston waterways.  
# Plot a bar graph in which x-axis is months (Jan-Dec) and y-axis is E.coli amount.  
# Each month is aggregated over the years. [25 points]


# extract the month from string format date, add new column named ‘month’
# take only the rows of that month: 
data['month'] = pd.DatetimeIndex(data['Date Collected']).month
date_01 = data.loc[data['month'] == 1]; date_01['E.coli']
date_02 = data.loc[data['month'] == 2]; date_02['E.coli']
date_03 = data.loc[data['month'] == 3]; date_03['E.coli']
date_04 = data.loc[data['month'] == 4]; date_04['E.coli']
date_05 = data.loc[data['month'] == 5]; date_05['E.coli']
date_06 = data.loc[data['month'] == 6]; date_06['E.coli']
date_07 = data.loc[data['month'] == 7]; date_07['E.coli']
date_08 = data.loc[data['month'] == 8]; date_08['E.coli']
date_09 = data.loc[data['month'] == 9]; date_09['E.coli']
date_10 = data.loc[data['month'] == 10]; date_10['E.coli']
date_11= data.loc[data['month'] == 11]; date_11['E.coli']
date_12= data.loc[data['month'] == 12]; date_12['E.coli']
```

```python
# mean of each month
mean_01= date_01['E.coli'].mean()
mean_02= date_02['E.coli'].mean()
mean_03= date_03['E.coli'].mean()
mean_04= date_04['E.coli'].mean()
mean_05= date_05['E.coli'].mean()
mean_06= date_06['E.coli'].mean()
mean_07= date_07['E.coli'].mean()
mean_08= date_08['E.coli'].mean()
mean_09= date_09['E.coli'].mean()
mean_10= date_10['E.coli'].mean()
mean_11= date_11['E.coli'].mean()
mean_12= date_12['E.coli'].mean()
```

```python
print('The average amount of E.coli per month in Houston waterways from January to December is \
respectively are:', round(mean_01, 3), round(mean_02,3), round(mean_03,3), round(mean_04,3),\
round(mean_05,3), round(mean_06,3), round(mean_07,3), round(mean_08, 3), round(mean_09,3),\
round(mean_10,3), round(mean_11,3),round(mean_12,3))
```

```python
# Plot a bar graph in which x-axis is months (Jan- Dec) and y-axis is E.coli amount. 

x = ('Jan',"Feb", "Mar", 'Ap', 'May', 'June','July','Aug','Sep','Oct','Nov','Dec')
y = [mean_01,mean_02,mean_03,mean_04,mean_05,mean_06,mean_07,mean_08,mean_09,mean_10,mean_11,mean_12]
plt.bar(x,y)
```

```python
# (3) Show correlations between E.coli and chemical factors such as water PH level, 
# dissolved oxygen, specific conductance, etc. [50 points]

# gather all the "chemical facotrs" columns to correlate with E.coli 
df =  data['E.coli']             
df1 = data['Dissolved Oxygen']
df2 = data['PH']
df3 = data['Salinity']
df4 = data['Ammonia Nitrogen']
df5 = data['Nitrate Nitrogen']
df6 = data['Nitrogen']
df7 = data['Phosphorous']
df8 = data['Chloride']
df9 = data['Sulfate']

# put into a nice data frame, define as "result"
frame = [df,df1,df2,df3,df4,df5,df6,df7,df8,df9]
result = pd.concat(frame, axis = 1)
result
```
